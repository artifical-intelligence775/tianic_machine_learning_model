```python
import pandas as pd 
file1 = "../input/titanic/train.csv"
df = pd.read_csv(file1)
```

# eliminating un-important variables(factors)

```python
df = df.drop(["PassengerId","Name","Embarked","Cabin","Ticket","Parch","SibSp"],axis = "columns")
df
```

# check if there are any NAN

```python
df.isnull().sum()
```

# use fillna to fill NAN

```python
import math 
median_age1 =math.floor(df.Age.median())
df.Age = df.Age.fillna(median_age1)
```

```python
df
```

```python
df.isnull().sum()
```

# converting str data to int

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df["Sex"] = le.fit_transform(df["Sex"])
```

```python
inputs = df.drop(["Survived"],axis = "columns")
target = df.Survived
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(inputs,target,test_size=0.2)
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```
